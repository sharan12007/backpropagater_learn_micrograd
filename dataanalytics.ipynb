```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import BinaryCrossentropy
import numpy as np
model=Sequential([
    Dense(units=25,activation='relu'),
    Dense(units=15,activation='relu'),
    Dense(units=1,activation='sigmoid')
])
model.compile(loss=BinaryCrossentropy)
model.summary()
X_train=np.random.rand(1000,10)
y_train=np.random.randint(2,size=(1000,1))
model.fit(x=X_train,y=y_train,epochs=100)
model.predict(X_train[:5])

```

```python
!pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt

# ReLU function
def relu(z):
    return np.maximum(0, z)

# Function to approximate parabola using multiple shifted ReLU neurons
def relu_parabola(x, num_neurons=5000):
    shifts = np.linspace(-3, 3, num_neurons)
    print(shifts)
      # activation points
    outputs = []
    for s in shifts:
        outputs.append(relu(x - s))  # each neuron activates at shift s
    outputs = np.array(outputs)

    # Use quadratic-like weights for approximation
    weights = -(shifts**2) + shifts.max()**2  
    weights = weights / np.max(np.abs(weights))  # normalize weights
    
    y_approx = np.dot(weights, outputs)
    return y_approx

# Input space
x = np.linspace(-3, 3, 400)
y_true = x**2
y_approx = relu_parabola(x, num_neurons=5000)

# Rescale approximation to match parabola range
y_approx = y_approx / np.max(y_approx) * np.max(y_true)

# Plot
plt.figure(figsize=(8,6))
plt.plot(x, y_true, label="True Function: $y=x^2$", linewidth=2, color="blue")
plt.plot(x, y_approx, label="ReLU Network Approximation", linestyle="--", linewidth=2, color="orange")
plt.legend()
plt.title("Approximating Parabola using ReLU Neurons")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
